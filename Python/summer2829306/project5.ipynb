```python
import os
import re
import json

class FileManager:
    def __init__(self,):
        pass
    
    def readFile(self,filename):
        cho = input("Read This file Answer Y/N ")
        if cho.lower() == 'y':
            print("_-_"*8,"ReadFile Menu","_-_"*8)
            print("1. Read Full File")
            print("2. Read first line ")
            print("3. read first 5 lines")
            print("4. Exit ReadFile")

            file = open(filename,'r')

            cho2 = int(input("selection : "))
            if cho2== 1:
                data = file.read()
                file.close()
            elif cho2 == 2:
                data = file.readline()
                file.close()
            elif cho2 == 3:
                data = file.readlines(5)
                file.close()
            else:
                print("Your Enter 4 Exiting ...")
                return
            return data
        else:
            return

    # file creation and reading RE based

    def searchFile(self):
        filename = input("Enter Filename + Extension to search for : ")
        print("^"*8,"SearchFile Menu","^"*8)
        print(f"1. List Directories in Current Directory")
        print(f"2. List all the Directories in the above Directory")
        print(f"3. Search {filename} in Current Directory")
        print(f"4. Search {filename} in Parent Directory")
        print(f"5. Enter Directory Path To Search for {filename} In")

        selection = int(input("Selection : "))
        if selection == 1:
            print(f"All the Directories in Current Directory\n {os.listdir(os.getcwd())}")
        elif selection == 2:
            parentDirectory = os.path.abspath("..")
            print(f"All the Directories in Directory {os.getcwd()} \n {os.listdir(os.getcwd())}")
        elif selection == 3:
            print(f"Searching in Current Working Directory : \n {os.getcwd()}")
            if os.path.exists(filename):
                print(f"{filename} Found")
                data = self.readFile(filename)
                if len(data) > 0:
                    print(data)
            else:
                print(f"File Not Found in {os.getcwd()}")
        elif selection == 4:
            parentDirectory = os.path.abspath("..")
            os.chdir(parentDirectory)
            print(f"Searching in {os.getcwd()}")

            if os.path.exists(filename):
                print(f"{filename} Found")
                data = self.readFile(filename)
                if len(data) > 0:
                    print(data)
            else:
                print(f"{filename} not Found in {os.getcwd()}")
        elif selection == 5:
            newDirectory = input("Enter Directory : ")
            os.chdir(newDirectory)
            print(f"Searcing in Directory {os.getcwd()}")
            if os.path.exists(filename):
                print(f"{filename} Found")
                data = self.readFile(filename)
                if len(data) > 0:
                    print(data)
            else:
                print(f"{filename} Not Found")
        else:
            print("Invalid Choice")

    def writeToFile(self,data):
        print("`"*10,"WriteToFile Menu","`"*10)
        print("1. Write to a file in Current Directory")
        print("2. Write to a file in Parent Directory")
        print("3. Enter Directory You Want to Write File To ")

        userchoice = int(input("User Choice : "))

        if userchoice == 1:
            filename = input("Enter File Name : ")
            if filename[-4:] == "json":
                with open(filename,'r')as file:
                    data = json.dumps(data)
                    file.write(data)
            else:
                with open(filename,'w+') as f:
                    f.write(data)
        elif userchoice == 2:
            filename = input("Enter File Name : ")
            parentDirectory = os.path.abspath("..")
            os.chdir(parentDirectory)

            if filename[-4:] == "json":
                with open(filename,'w+') as f:
                    f.write(json.dumps(data))
            else:
                with open(filename,'w+') as f:
                    f.write(data)
        elif userchoice == 3:
            inputDirectory = input("Enter Directory Full Path : ")
            filename = input("Enter File Name + Extension : ")

            os.chdir(inputDirectory)

            if filename[-4:] == "json":
                with open(filename,'w+') as f:
                    f.write(json.dumps(data))
            else:
                with open(filename,'w') as f:
                    f.write(data)    
            

    # asking user to search for email , phone in the read data

    def searchForContent(self,data):
        print("--"*8,"SearchForContent Menu","--"*8)
        print("1. Search for Emails : ")
        print("2. Search for phone numbers : ")
        print("3. Search for IP Addresses : ")

        userchoice = int(input("Your Choice : "))

        if userchoice == 1:
            emailPattern  = r"\S+\.?\S*@\S+\.\S+\.?\S*"
            emailResult = re.findall(emailPattern,data)
            return emailResult
        
        elif userchoice == 2:

            phonePatterns = [r"\d{11}",r"\+\d+\s*\d+\s*\d+"]
            phoneResult = []
            for pattern in phonePatterns:
                phoneResult.append(re.findall(pattern,data))
            return phoneResult
        elif userchoice == 3:
            ip_pattern = r"\d+\.\d+\.\d+\.\d+"
            ip_address = re.findall(ip_pattern,data)
            return ip_address
        else:
            print("Invalid Choice Returning....")
            return
        


fm1 = FileManager()

if __name__ == "__main__":

    while True:
        print("⋙"*8, "fILeMnaGeR FiaQ","⋘"*8)
        print("1. Search and Read  File")
        print("2. Write to a File")
        print("3. Read A file")
        print("4. Search for content in a file")
        print("5. Exit")


        userchoice = int(input("Enter you choice : "))

        if userchoice == 1:
            fm1.searchFile()# make sure you enter the file name in the input

        elif userchoice == 2:
            # make sure you give the data as input
            data = fm1.readFile("hola.txt")
            fm1.writeToFile(data)
        elif userchoice == 3:
            filenamee = input("Enter filename + extension")
            readfiledata = fm1.readFile(filenamee)
            print(readFile)
        elif userchoice == 4:
            file_noma = input("Enter File Name")
            data = fm1.readFile(file_noma)
            result = fm1.searchForContent(data)
            print(result)
        else:
            print("thank you for using File Manager fAIq")
            break
```
