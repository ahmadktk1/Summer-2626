```python
import random
import json
class Hostel:
    def __init__(self):
        pass

    def savefile(self,filename,extension,data):
        """
        __Summary__:
            This file saves data in given file 
        Parameter:
            filename:
                    filename + extension e.g data.json , journel.txt
            data:
                    the data you want to save in file
        """
        if extension.lower() == "json":
            with open(f"{filename}.{extension}","w") as f:
                data = json.dumps(data)
                f.write(data)
        else:
            with open(f"{filename}.{extension}",'w') as f:
                f.write(data)
        return "File Saved Successfully"


    def loadfile(self,filename,extension):
        """
        __Summary__:
            This file load data from given file and store it in variable
        Parameters:
            filename:
                fiel name + its extension
        """
        with open(f"{filename}.{extension}", 'r') as f:
            if extension == "json":
                data = json.load(f)
            else:
                data = f.read()
        return data
    
    def hostelite(self,hostelers,accounts):
        name = input("Name : ")
        fname = input("Father Name : ")
        phone = input("Phone Number: ")
        mail = input("Email Address : ")
        student = True
        student = input("True/ False").capitalize()
        institute = input("Institute Name : ")
    
        # we will implement the exception handling thing but letter
        # for now lets focus on generating the username and pin for a user 
        # username = firstname + last 3 digits of phone 
        # pin randomly generated 5 digits
    
        
        UID = accounts.keys()
        UID = len(UID)
        UID = UID + 1
    
    
        # creating a username 
        try:
            first,last = name.split(" ")
        except ValueError:
            first = name
        
        username = first+phone[-3:]
        pin = random.randint(11111,99999)
    
        user = {
            "UserDetails":[name,fname,phone,mail,student,institute],
            "Account":[username,pin]
        }
    
        accounts[UID] = [username,pin]
        hostelers[UID] = user
    
        print("Successfully Added")
        
    def showMenu(self,menu:dict):
        # this function will show the weekly Menu which will be stored in Menu.json
        print(f"Day \t\t Breakfast \t\t Afternoon\t\t Dinner")
        # keyys = []
        for i in menu.keys():
            # for j in menu[i].key():
            #     keyys.append(i)
            print(f"{i}>---------<{menu[i]['Breakfast']}>---------<{menu[i]['Noon']}>---------<{menu[i]['Dinner']} ")
    

    def initializeRooms(self,):
        rooms = {}
        for i in range(17):
            kamra=  {}
            
            for j in range(4):
                kamra[j+1] = ["",""]
        
            rooms[f"room{i+1}"] = kamra

        self.savefile("roomsSeating",'json',rooms)
    # creating a room allocation and room seating function

    # this will contain the name and room of the person and His mail Id

    # how we will do it lets us see
    # this function will first lets us see all the rooms seating then If we want to change it

    def showSeating(self,roomSeating):
        for key in roomSeating.keys():
            for seat in roomSeating[key].keys():
                print(f"{key}\t Seat {seat} Name  {roomSeating[key][seat][0]}  Phone {roomSeating[key][seat][1]}  ")
   
    def allocateRoom(self,rooms):
        name = input("Name : ").capitalize()
        phone = input("Phone : ")
        room = input("room: format(room1,room5) : ").lower()
        avail = []
        for key in rooms[room].keys():
            
            if len(rooms[room][key][0]) == 0:
                avail.append(key)
        
        if len(avail) == 0:
            raise ValueError(f"{room} is full")
        else:
            for i in avail:
                rooms[room][i] = [name,phone]
                print(f"{name} is allocated to {room} seat {i}  ")
                break


    def deallocateRoom(self,rooms):
        name = input("name : ").capitalize()
        room = input("room1 : ").lower()
        found = False
        for seat in rooms[room].keys():
            if name == rooms[room][seat][0]:
                found = True
                rooms[room][seat] = ["",""]
        
        if not found:
            return f"{name} not found"
        else:
            return f"Successfully de allocated the seat"
        

    def roomAllocation(self,roomsSeating):
        # the function should Initialy show the rooms seating and then ask user to update
        # I think I have to ask user to choose what todo and use while loop
        # if there is already a hosteler it shall raise an error
        while True:
            print("1. See Rooms Seating")
            print("2. Allocate New Hosteler to Room")
            print("3. Deallocate Hosteller Room")
            print("4. Finish Room Allocation Menu")
            choice = int(input("Enter Choice : "))

            if choice == 1:
                self.showSeating(roomsSeating)
            elif choice == 2:
                # there is a way to show which rooms are availble but this will the code lengthy
                # simple entry 
                self.allocateRoom(roomsSeating)
            elif choice == 3:
                self.deallocateRoom(roomsSeating)
            else:
                print("Exiting Room Allocation Menu . . .")
                break
                
    def initializedFeeRecord(self,feeRecord,rooms):
        for key in rooms.keys():
            for seat in rooms[key].keys():
                if [rooms[key][seat][0],key] in feeRecord["paid"]:

                    pass
                elif  [rooms[key][seat][0],key] in feeRecord['unpaid']:
                    pass
                else:
                    feeRecord["unpaid"].append([rooms[key][seat][0],key])

    def trackFeeRecord(self,feeRecord):
        print(f"Name    Room        Status")
        for key in feeRecord.keys():
            for i in feeRecord[key]:
                    # print(feeRecord[key][0])
                # print(f"{feeRecord[key][i][0]}    {feeRecord[key][i][1]}     {key}")
                print(f"{i[0]} {i[1]} {key}")



    # now the last function pays fees
    def pay(self,feeRecord,paymentRecord:list):
        name = input("Name : ").capitalize()
        roomno = input("Room No (room1): ").capitalize()

        if [name,roomno] in feeRecord["paid"]:
            return f"Already Paid"
        elif [name,roomno] in feeRecord["unpaid"]:
            print("Send your fees to this Easypiasa Account")
            print("Account No.   03454545345")
            print("Account Holder Name : Latamber Kali Janan")
            transId = input("Enter transection ID")
            del feeRecord["unpaid"][name,roomno]
            feeRecord["paid"].append([name,roomno])
            paymentRecord.append([name,roomno,transId])
        
```
