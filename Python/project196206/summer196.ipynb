```python
def searchBook(books=books):
    book = input("Book Name : ")
    result = 0
    for i,j in books.items():
        if j[1] != book:
            result = "Book not found"
            break
        elif j[1] == book:
            result = books[i]
            break
    return result

searchBook()
```
